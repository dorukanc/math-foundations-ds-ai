# sample mean converges to population mean when sample grows

```python
# Import binom and describe

# the prob of exactly k success in n trials

from scipy.stats import binom
from scipy.stats import describe

# Sample of 250 fair coin flips

samples = binom.rvs(n=1, p=0.5, size=250, random_state = 42)
```

```python
# print the first 100 values from the sample

print(samples[0:100])
```

```python
# Calculate the sample mean

print(describe(samples[0:10]).mean)
```

```python
# Plotting the sample mean

import matplotlib.pyplot as plt

# Define our variables

coin_flips, p, sample_size, averages = 1, 0.5, 1000, []
```

```python
# Generate the sample

sample2 = binom.rvs(n=coin_flips, p=p, size = sample_size, random_state=42)
```

```python
# Calculate the sample mean

for i in range(2, sample_size+1):
    averages.append(describe(sample2[0:i]).mean)

# Print the first values of averages
print(averages[0:10])
```
