```python
# Poisson population plot

from scipy.stats import poisson, describe
from matplotlib import pyplot as plt
import numpy as np
```

# Central Limit Theorem (CLT): Sum of independent, identically distrubuted random variables tends to a normal dist as the number of variables increases. This holds true even if the original variables are not normally distrubuted.

```python
# Generate our population 

population = poisson.rvs(mu=2, size=1000, random_state=20)
```

```python
# Draw the histogram with labels

plt.hist(population, bins = range(9), width=0.8)
plt.show()
```

```python
# Sample means plot
# generate 350 sample means, selecting from pop values

np.random.seed(42) # set random seed

# Define list of sample means
sample_means = []

for _ in range(350):
    # Select 10 from pop
    sample = np.random.choice(population, 10)
    # calculate sample mean of sample
    sample_means.append(describe(sample).mean)
```

```python
# Draw the histogram with labels

plt.xlabel("Sample mean values")
plt.ylabel("Frequency")
plt.title("Sample means histogram")
plt.hist(sample_means)
plt.show()
```
